```python
#Hasta acá se ve que una caida genera un peak de al menos dos variables en un lapso muy breve de tiempo.
#Hay variables que sin presentarse una caida generan peak pero con duración mayor y sin otro comportamiento similar o de igual magnitud en otra variable.
#cargar libreria isolation forest
from sklearn.ensemble import IsolationForest
```

```python
#Isolation forest para Exp1F y variable peak "RotXaxis"
isolation_forest = IsolationForest(n_estimators=100)

isolation_forest.fit(Exp1F['RotXaxis'].values.reshape(-1, 1))

xx = np.linspace(Exp1F['RotXaxis'].min(), Exp1F['RotXaxis'].max(), len(Exp1F)).reshape(-1,1)

anomaly_score = isolation_forest.decision_function(xx)

outlier = isolation_forest.predict(xx)

plt.figure(figsize=(10,4))

plt.plot(xx, anomaly_score, label='anomaly score')

plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score), 

                 where=outlier==-1, color='r', 

                 alpha=.4, label='outlier region')

plt.legend()

plt.ylabel('anomaly score')

plt.xlabel('RotXaxis')

plt.show();
```

```python
from sklearn.preprocessing import LabelEncoder
#cambiar de esta forma los labels en los datos Exp de data
cols=['AccelXaxis','AccelYaxis','AccelZaxis','RotXaxis','RotYaxis','RotZaxis','OtherAccelXaxis','OtherAccelYaxis','OtherAccelZaxis']
df = pd.read_csv("data/F01_SA01_R01(1).txt", sep=",", names=cols + ["label"], index_col=None)
df.shape
encs = dict()
data = df.copy()  #.sample(frac=1)
for c in data.columns:
    if data[c].dtype == "object":
        encs[c] = LabelEncoder()
        data[c] = encs[c].fit_transform(data[c])
df.head()
```

```python
#isolation forest multivariable (RotXaxis; AccelXaxis)

X_train, y_train = data[cols][:1500], data["label"][:1500].values
X_valid, y_valid = data[cols][1500:2400], data["label"][1500:2400].values
X_test, y_test = data[cols][2400:], data["label"][2400:].values
```

```python
iForest = IsolationForest(n_estimators=100, max_samples=256, contamination=0.2, random_state=2018)
iForest.fit(X_train)
scores = iForest.decision_function(X_valid)
plt.figure(figsize=(12, 8))
plt.hist(scores, bins=50);
```

```python
from sklearn.metrics import roc_auc_score
y_pred_train = iForest.predict(X_train)
y_pred_dev = iForest.predict(X_valid)
y_pred_test = iForest.predict(X_test)

print("Accuracy dev :", list(y_pred_dev).count(1)/y_pred_dev.shape[0])
print("Accuracy test:", list(y_pred_test).count(-1)/y_pred_test.shape[0])
#print("AUC: {:.1%}".format(roc_auc_score((-0.2 < scores), y_valid == list(encs["label"].classes_).index("normal."))))
```

```python
scores_test = iForest.decision_function(X_test)
print("AUC: {:.1%}".format(roc_auc_score((-0.2 < scores_test), y_test == list(encs["label"].classes_).index("normal."))))
```
